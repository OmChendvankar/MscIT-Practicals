```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
import os
import pandas as pd
```

```python
InputFileName='Retrieve_Router_Location.csv'
InputFileDir = 'C:/VKHCG/01-Vermeulen/01-Retrieve/01-EDS/02-Python'
OutputFileDir = 'C:/VKHCG/01-Vermeulen/02-Assess/01-EDS/02-Python'
OutputFileName='Assess-DAG-Company-GPS.png'
```

```python
sFileName= InputFileDir + '/' + InputFileName
print('Loading :',sFileName)
```

```python
CompanyData=pd.read_csv(sFileName,header=0,low_memory=False, encoding="latin-1")
print('Loaded Company :',CompanyData.columns.values)
```

```python
CompanyData.shape
```

```python
G=nx.Graph()
print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
for i in range(CompanyData.shape[0]):
    nLatitude=round(CompanyData['Latitude'][i],1)
    nLongitude=round(CompanyData['Longitude'][i],1)
    
    if nLatitude < 0:
        sLatitude = str(nLatitude*-1) + ' S'
    else:
        sLatitude = str(nLatitude) + ' N'
        
    if nLongitude < 0:
        sLongitude = str(nLongitude*-1) + ' W'
    else:
        sLongitude = str(nLongitude) + ' E'
        
    sGPS= sLatitude + '-' + sLongitude
    G.add_node(sGPS)
```

```python
for n1 in G.nodes():
    for n2 in G.nodes():
        if n1 != n2:
            print('Link :',n1,' to ', n2)
            G.add_edge(n1,n2)
```

```python
print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
sFileName=OutputFileDir + '/' + OutputFileName
print('Storing :', sFileName)
```

```python
pos=nx.circular_layout(G,dim=2, scale=2)
nx.draw(G,pos=pos, 
        nodecolor='r',edge_color='b',
        with_labels=True,node_size=4000,
        font_size=9)
plt.savefig(sFileName) # save as png
plt.show() # display
```
